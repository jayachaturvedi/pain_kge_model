queries on neo4j

```python
# query - Find a concept based on an ID.

query = """
MATCH (n:Concepts {id: 'C0030193'})
	RETURN n.id, n.name
"""

df_pain_concept = graph.run(query).to_data_frame()
df_pain_concept
```

```python
# query - Find 1-order (or direct) parent nodes of a concept along a relation

query = """
Match (c:Concepts{id: 'C0030193'})-[:cui_2_cui{relation:'isa'}]->(p:Concepts)
RETURN p.id, p.name
"""

df_pain_parent = graph.run(query).to_data_frame()
df_pain_parent
```

```python
# query - Find 1-order (direct) child nodes of a concept along a relation

query = """
Match (p:Concepts{id:'C0030193'})<-[:cui_2_cui{relation:'isa'}]-(c:Concepts)
RETURN c.id, c.name
"""

df_pain_child = graph.run(query).to_data_frame()
df_pain_child['c.name'] = df_pain_child['c.name'].str.lower()
df_pain_child
```

```python
df_pain_child.to_csv('./umls_pain_child_list.csv', index=False)
```

```python
# query - Find all child (n-order) nodes of a concept along a relation - this example is for relation 'is-a'

query = """
MATCH (p:Concepts{id:'C0030193'})<-[r:cui_2_cui*0..{relation:'isa'}]-(c:Concepts)
return c
"""

df_pain_child_relation = graph.run(query).to_data_frame()
df_pain_child_relation
```

```python
df_pain_child_relation['c'][5]
```

```python
# query - Find all parent (n-order) of a concept along a relation - this example is for relation 'is-a'

query = """
MATCH (c:Concepts{id:'C0030193'})-[r:cui_2_cui*0..{relation:'isa'}]->(p:Concepts)
return p
"""

df_pain_parent_relation = graph.run(query).to_data_frame()
df_pain_parent_relation
```

```python
df_pain_parent_relation['p'][3]
```

```python
# query - Find the names of all unique parent nodes of a concept

query = """
MATCH (c:Concepts{id:'C0030193'})<-[r:cui_2_cui*0..{relation:'isa'}]-(p:Concepts)
return c.name, collect( p.name)
"""

df_pain_parent_unique = graph.run(query).to_data_frame()
df_pain_parent_unique
```

```python
# query - Find an ICD-9 code (pain is 338) - how can i list all that fall under 338? like 338.29

query = """
MATCH (sc: Terms {other_code: '338'}) 
RETURN sc;
"""

df_pain_icd9 = graph.run(query).to_data_frame()
df_pain_icd9['sc'][0]
```

```python
# query - Find a SNOMED code - how can i get child nodes?

query = """
MATCH (sc: Terms {code:'22253000'}) 
RETURN sc;
"""

df_pain_snomed = graph.run(query).to_data_frame()
df_pain_snomed['sc'][0]
```
